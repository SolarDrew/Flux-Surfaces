```python
import yt
import pysac
import pysac.yt
import matplotlib.pyplot as plt
```

```python
#thisfile = yt.load('/fastdata/sm1ajl/Flux-Surfaces/gdf/test_p120-0_0-5_0-5/test_p120-0_0-5_0-5_00001.gdf')
thisfile = yt.load('/fastdata/sm1ajl/Flux-Surfaces/gdf/m0_p120-0_0-5_0-5/m0_p120-0_0-5_0-5_00001.gdf')
```

```python
myplot = yt.SlicePlot(thisfile, 'x', 'density_pert', axes_unit='Mm')
#myplot.set_cmap('density_bg', 'viridis')
myplot.set_cmap('density_pert', 'coolwarm')
myplot.show()
```

```python
myplot = yt.SlicePlot(thisfile, 'z', 'density_pert', axes_unit='Mm')
#myplot.set_cmap('density_bg', 'viridis')
myplot.set_cmap('density_pert', 'viridis')
myplot.show()
```

```python
for field in thisfile.field_list:
    myplot = yt.SlicePlot(thisfile, 'x', field, axes_unit='Mm')
    myplot.set_cmap(field, 'viridis')
    myplot.show()
```
